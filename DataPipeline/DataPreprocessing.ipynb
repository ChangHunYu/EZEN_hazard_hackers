```python
import pandas as pd
import json
import requests
```

# 대륙 저장

```python
pd.DataFrame(alertHistory)["continent_eng_nm"].value_counts()
```

```python
pd.DataFrame(alertHistory)["continent_nm"].value_counts()
```

```python
unique_continent_eng_nm  = pd.DataFrame(alertHistory)['continent_eng_nm'].unique()
```

```python
unique_continent_nm  = pd.DataFrame(alertHistory)['continent_nm'].unique()
```

```python
result = [
    { "continentEngNm": eng, "continentNm": nm }
    for eng in unique_continent_eng_nm
    for nm in unique_continent_nm
    if eng == unique_continent_eng_nm[unique_continent_nm.tolist().index(nm)]
]
# 중복 제거
ContinentRequestList = [dict(t) for t in {tuple(d.items()) for d in result}]

print(ContinentRequestList)
```

```python
for row in ContinentRequestList:
    requests.post("http://localhost:8080/continent", json=row )
```

# 국가 저장

```python
with open('CountryAlert.json', 'r', encoding='UTF-8') as f:
    countryAlert = json.load(f)
pd.DataFrame(countryAlert).value_counts()
```

```python
country_search_japan = pd.DataFrame(countryAlert)[pd.DataFrame(countryAlert)['country_eng_nm'] == 'Japan']
pd.DataFrame(country_search_japan)
```

```python
countryAlert = pd.DataFrame(countryAlert)
# written_dt 열을 datetime 형식으로 변환
countryAlert['written_dt'] = pd.to_datetime(countryAlert['written_dt'])
# written_dt 열을 기준으로 오름차순 정렬
sorted_countryAlert = countryAlert.sort_values(by='written_dt')

sorted_countryAlert = pd.DataFrame(sorted_countryAlert)
sorted_countryAlert
```

```python
CountryRequestList = []
response = requests.get('http://localhost:8080/continent')
continent_list = response.json()

for index, row in sorted_countryAlert.iterrows():
    continentId = next((continent['id'] for continent in continent_list if continent['continentEngNm'] == row['continent_eng_nm']), None)
    countryEngName = row['country_eng_nm']
    countryIsoAlp2 = row['country_iso_alp2']
    countryName = row['country_nm']
    flagDownloadUrl = row['flag_download_url']
    mapDownloadUrl = row['map_download_url']
    
    
    CountryRequestList.append({
        "continentId" : continentId,
        "countryEngName": countryEngName,
        "countryIsoAlp2" : countryIsoAlp2,
        "countryName" : countryName,
        "flagDownloadUrl" : flagDownloadUrl,
        "mapDownloadUrl": mapDownloadUrl,
    })
```

```python
pd.DataFrame(CountryRequestList).value_counts()
```

```python
for row in CountryRequestList:
    response=requests.post("http://localhost:8080/country", json=row )
    print(response.text)
```

# 여행경보 저장

```python
AlertRequestList = []
for index, row in sorted_countryAlert.iterrows():
    # "alarm_lvl"이 NaN일 경우 0으로 대체
    level = int(row['alarm_lvl']) if not pd.isna(row['alarm_lvl']) else 0
    regionType = row['region_ty']
    remark = row['remark']
    dang_map_download_url = row['dang_map_download_url']
    written_dt = row['written_dt']
    countryEngName = row['country_eng_nm']
    
    
    AlertRequestList.append({
        "countryEngName" : countryEngName,
        "level": level,
        "message": "no message",
        "description" : "no description",
        "regionType" : regionType,
        "remark" : remark,
        "dang_map_download_url": dang_map_download_url,
        "written_dt": written_dt
    })


```

```python
pd.DataFrame(AlertRequestList).value_counts()
# print(AlertRequestList)
```

```python
alertHistory = pd.DataFrame(alertHistory)
# written_dt 열을 datetime 형식으로 변환
alertHistory['wrt_dt'] = pd.to_datetime(alertHistory['wrt_dt'])
# written_dt 열을 기준으로 오름차순 정렬
sorted_alertHistory = alertHistory.sort_values(by='wrt_dt')

sorted_alertHistory = pd.DataFrame(sorted_alertHistory)
sorted_alertHistory
```

```python
japan_alert_history = sorted_alertHistory[sorted_alertHistory['country_eng_nm'] == 'Japan']
pd.DataFrame(japan_alert_history)
```

```python
for index, row in sorted_alertHistory.iterrows():
    country = row['country_eng_nm']
    matching_requests = [item for item in AlertRequestList if item['countryEngName'] == country]
    if matching_requests[0]['written_dt'] == row['wrt_dt']:
        newItem = matching_requests[0]
        newItem['message'] = row['title']
        newItem['description'] = row['txt_origin_cn']
    if matching_requests[0]['written_dt'] != row['wrt_dt']:
        newItem = matching_requests[0].copy()
        newItem['message'] = row['title']
        newItem['description'] = row['txt_origin_cn']
        newItem['written_dt'] = row['wrt_dt']
        AlertRequestList.append(newItem)
```

```python
pd.DataFrame(AlertRequestList).value_counts()
```

```python
for row in AlertRequestList:
    # print(row)
    if isinstance(row.get('written_dt'), pd.Timestamp):
        row['written_dt'] = row['written_dt'].isoformat()
    response=requests.post("http://localhost:8080/alerts", json=row )
    print(response.json())
```
